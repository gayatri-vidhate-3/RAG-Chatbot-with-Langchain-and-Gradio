## Extract data from PDF : Document loader

```python
# from langchain.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/Gayatri_Vidhate_CV.pdf")

text = loader.load()
```

```python
text[0]
```

```python
text_content = ""

for t in text:
  text_content += t.page_content
```

```python
text_content
```

```python
groq_model.invoke(f"""Based on given context answer the question.
Context : {text_content}
question : What are the different projects Gayatri has done in ZIGRAM?
""")
```

## Chunking

```python
text
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=300,           # chunk_size : No. of characters in single chunk

                                               chunk_overlap=100)         # chunk_overlap : overlap between two consecutive chunks

chunks = text_splitter.split_documents(text)
```

```python
chunks
```

## Embedding Model

```python
!pip install -U sentence-transformers langchain-huggingface
```

```python
# from langchain_openai import OpenAIEmbeddings

# embedding_model = OpenAIEmbeddings()          # Ada

# because of limited access of OpenAI API Key using Huggingface Embeddings

from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
embedding_model
```

# Vector DB

```python
!pip install chromadb
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(documents=chunks,
                           embedding=embedding_model
                           )
```

## Retriever

```python
context_retriever = db.as_retriever()
```

```python
context_retriever.invoke("How much is the Research grant received?")
```

## Chain

```python
import langchain
import langchain_community

print("LangChain:", langchain.__version__)
print("LangChain Community:", langchain_community.__version__)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

prompt = ChatPromptTemplate.from_template(
    "Answer the question based only on the context below:\n\n{context}\n\nQuestion: {input}"
)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context": context_retriever | format_docs, "input": RunnablePassthrough()}
    | prompt
    | groq_model
    | StrOutputParser()
)

response = rag_chain.invoke("What is the gihub link address")
print(response)
```
